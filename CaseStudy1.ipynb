```python
dfp = sns.load_dataset("titanic")
dfp.head()
```

```python
## number of male and female passenger
dfp["sex"].value_counts()
```

```python
## uniq number of for each columns
dfp.nunique()
```

```python
## uniq number of "pclass" value
dfp["pclass"].nunique()
```

```python
## what type the "embark" values and changed type
dfp["embarked"].dtype
dfp["embarked"] = dfp["embarked"].astype("category")
dfp["embarked"].dtype
```

```python
## All information of those with embarked value c
dfp[dfp["embarked"] == "C"].head(10)
```

```python
## information for passenger who are under 30 age and female

dfp[(dfp["age"] < 30) & (dfp["sex"] == "female")].head(10)
```

```python
## delete the who value from the dataframe
dfp.drop("who", axis=1, inplace=True)
dfp.head(10)
```

```python
## most repeated values in deck variable
dfp["deck"].mode()[0]
```

```python
## count the null values in deck variable
dfp["deck"].isnull().sum()
```

```python
## replacing null values in deck with the most repeated
```

```python
dfp["deck"].fillna(dfp["deck"].mode()[0], inplace=True)
```

```python
## again count the null values in deck variable
dfp["deck"].isnull().sum()
```

```python
## sum count mean values of survived variable in pclass and sex breakdown

dfp.groupby(["pclass","sex"]).agg({"survived":["sum","count","mean"]})
```

```python
dfl = sns.load_dataset("tips")
dfl.head()
```

```python
## Find the sum, min , max and average of total bill and tips values of lunch time and female customers accordig to day
dfl[(dfl["time"] == "Lunch") & (dfl["sex"] == "Female")].groupby("day").agg({"total_bill": ["sum","min","max","mean"],
                                                                          "tip":["sum","min","max","mean"]})
```
