```python
## Kategorik değişkenler: sütun grafiği - countplot bar -
## sayısal değişkenler - hist, boxplot -
```

```python
import pandas as pd
import seaborn as sns
```

```python
pd.set_option("display.max_columns", None)
df = sns.load_dataset("titanic")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
## kategorik değişken görselleştirme
df["sex"].value_counts().plot(kind="bar")
plt.show()
```

```python
## sayısal değişken görselleştirme
plt.hist(df["age"])
plt.show()
```

```python
plt.boxplot(df["fare"])
plt.show()
```

```python
dfs = sns.load_dataset("tips")
dfs.head()
```

```python
sns.countplot(x=dfs["sex"], data=dfs)
plt.show()
```

```python
sns.boxplot(x=dfs["total_bill"], data=dfs)
plt.show()
```
